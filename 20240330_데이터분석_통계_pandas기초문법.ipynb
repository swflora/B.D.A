```python
#numpy 내에서도 기초통계치를 계산하는 함수 존재
#statistics 기초통계치 계산하는 함수
# scipy 등 다양하게 있다. (실제 판다스로도 계산 가능하다.)
import numpy as np
import statistics
```

## 데이터의 다양한 요약지표 중
### 데이터의 대푯값이라고 할 수 있는 평균외에 다양한 지표들 문법
- 평균을 구하는 함수
- 중위값 구하는 함수
- 분산
- 표준편차
- 최대,최소
- 범위수 (4분위수)
- 상관계수

```python
data = np.array([1,2,4,5,6,7,8,9,10,10])

#넘파이의 다양한 문법
print(np.mean(data)) # 평균

print(np.median(data)) # 중앙값

print(np.std(data)) #표준편차

print(np.var(data)) # 분산

print(np.min(data)) #최솟값

print(np.max(data)) #최댓값

print(np.argmax(data)) # 최대값의 인덱스 반환

print(np.argmin(data)) # 최소값의 인덱스 반환

print(np.percentile(data, 25)) # 제 1사분위수 25%

print(np.percentile(data, 75)) # 제 3사분위수 75%
```

```python
#상관계수
data2 = np.array([12,13,14,15,16,17,18,19,20,20])
print(np.corrcoef(data, data2))
print('')

#Statistic 다양한 문법

print(statistics.mean(data))

print(statistics.median(data))

print(statistics.variance(data))

print(statistics.stdev(data))

print(statistics.harmonic_mean(data)) #다른 평균들 계산 가능

#최빈값
print(statistics.mode(data)) #가장 빈도가 많은 값을 찾아준다.

print(statistics.quantiles(data, n=4))
```

# 평균

# 중앙값
- 데이터를 크기 순서대로 나열할 때 정확하게 중앙에 위치한 값
- 홀수면 중앙이 나온다, 짝수면 두 개의 값을 평균으로 계산한다.

- 중앙값은 평균값에 비해서는 이상치에 영향을 덜 받는다.
- 평균으로 계산하는 게 아니라, 중앙값으로 대치해서 MSE 바꿔서 중앙값으로 바꿔서 계산하고 하는 이유가
- 데이터 자체에 이상치가 큰 경우 아무래도 대푯값을 평균보다는 중앙값이 적절하게 볼 수 있다. ( 도메인적인 부분과 같이 고려해야 한다.)

## 데이터의 산포도, 흩어진 정도를 바라보는
# 편차
- 편차는 각 데이터가 평균으로부터 얼마나 떨어져 있는지 지표
# 분산
- 산포도의 지표로 데이터 평균간의 차이를 이용, 분산은 편차의 제곱을 이용하여 계산
# 표준편차
- 산포도 지표로 분산에 제곱근을 취한 것을 표준편차

```python
#간단한 편차 계산
df_eng_mean=np.mean(df.english)

df['eng_dv']=df.english - df_eng_mean

# 편차와 같은 값이 나온다
print(np.mean(df['eng_dv']**2))
print(np.var(df.english))
```

# 사분위 범위, 범위

- 범위는 range
- 분산이나 표준편차와는 다르게 데이터를 전체를 보는 게 아니다.
- 최댓값, 최솟값으로만 산포도를 표현하는 것
- 최댓값과 최솟값이 차이가 크면 산포도가 크고, 차이가 작으면 작다.
- Rg = X.max - X.min

```python
np.max(df.english)- np.min(df.english)
```

- 사분위 범위
- 범위도 결국, 최대, 최소로 계산하는 것 -> 이상치가 크면 변화가 크다.
- 사분위 범위 interquartile range
- 데이터를 1로 보면 하위 25%, 중위 50%, 상위 75%
- 제1사분위수, 제2사분위수(50%), 제3사분위수 Q1, Q2, Q3
- 사분위의 범위 IQR
- IQR = Q3- Q1
