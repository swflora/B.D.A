# 남성의 생존확률 구하기

```python
import seaborn as sns
tt = sns.load_dataset('titanic')

tt
```

```python
# 클래스 별 생존 확률
print(tt.query("""pclass == 3""").shape[0])
print(tt.query("""pclass == 2""").shape[0])
print(tt.query("""pclass == 1""").shape[0])
```

```python
# 클래스 별로 생존 남성의 수 구하기
tt['pclass_prob'] = 0

for i in range(len(tt)):
    if tt.loc[i]['pclass'] == 3 and tt.loc[i]['sex'] == 'male':
        tt['pclass_prob'].loc[i] = tt.loc[i]['survived'].sum() / 491
    elif tt.loc[i]['pclass'] == 2 and tt.loc[i]['sex'] == 'male':
        tt['pclass_prob'].loc[i] = tt.loc[i]['survived'].sum() / 184
    elif tt.loc[i]['pclass'] == 1 and tt.loc[i]['sex'] == 'male':
        tt['pclass_prob'].loc[i] = tt.loc[i]['survived'].sum() / 216

for i in range(len(tt)):
    if tt.loc[i]['pclass'] == 3 and tt.loc[i]['sex'] == 'female':
        tt['pclass_prob'].loc[i] = tt.loc[i]['survived'].sum() / 491
    elif tt.loc[i]['pclass'] == 2 and tt.loc[i]['sex'] == 'female':
        tt['pclass_prob'].loc[i] = tt.loc[i]['survived'].sum() / 184
    elif tt.loc[i]['pclass'] == 1 and tt.loc[i]['sex'] == 'female':
        tt['pclass_prob'].loc[i] = tt.loc[i]['survived'].sum() / 216

```
